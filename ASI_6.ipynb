```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
from sklearn.model_selection import train_test_split
```

```python
import numpy
```

```python
data_url = 'https://raw.githubusercontent.com/michaljn2/ASI_Projekt/ASI_5/data/'
all_data = TabularDataset(f'{data_url}weatherAUS.csv')


all_data.head()
```

```python
label = "RainTomorrow"
```

```python
all_data.dropna(axis=0, subset=['RainTomorrow'], inplace=True)
```

```python
train_data, test_data = train_test_split(all_data, test_size=0.2, random_state=100)
```

```python

train_data[label].describe()
```

```python
predictor = TabularPredictor(label=label).fit(train_data, time_limit=60)
```

```python
y_pred = predictor.predict(test_data.drop(columns=[label]))
y_pred.head()
```

```python
predictor.evaluate(test_data, silent=True)
```

```python
predictor.leaderboard(test_data, silent=True)
```
